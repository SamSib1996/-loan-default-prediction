```python
#setting up data 
#getting ready for random cloassification model 
```

```python
loan_data.info()
```

```python
# categorical features - the purpose column is categorical - change using dummy variable so that sklearn can understand them 
cat_feats=['purpose']
```

```python
#Now use pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data=pd.get_dummies(loan_data,columns=cat_feats,drop_first='true')
```

```python
final_data.info()
```

```python
# Train Test Split 
# now let us split our data into training set and testing set 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= final_data.drop('not.fully.paid',axis =1)
y= final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Training a decision tree Model 
#let us train single decision tree first 
from sklearn.tree import DecisionTreeClassifier
```

```python
#creating an instance of DecisionTreeClassifier calld dtree and fit it to training data 
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
# Predictions and evaluation of decision tree
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
# Training random forest model 
from sklearn.ensemble import RandomForestClassifier 
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

```python
# prediction and Evaluation 
predictions =rfc.predict(X_test)
```

```python
# classification report from the results
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
# Showing the Confusion Matrix for the predictions
print(confusion_matrix(y_test,predictions))
```
